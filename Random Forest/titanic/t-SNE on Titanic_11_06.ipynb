```python
from pandas import read_csv
import pandas as pd

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df1=pd.get_dummies(df, columns=['Sex','Embarked','Pclass']).drop('Cabin', axis=1).fillna(df['Age'].mean())
```

```python
#this is for train data
df2=df1[['Age', 'Parch','Embarked_Q', 'SibSp','Fare',
        'Embarked_C', 'Embarked_S', 'Pclass_1', 'Pclass_2','Sex_male']]
```

```python
df2
```

```python
from bioinfokit.visuz import cluster
data_tsne = TSNE(n_components=2).fit_transform(df2)
cluster.tsneplot(score=data_tsne)
```

```python
# get a list of categories
color_class = df['Survived']
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right')

#Plot will be stored in the default directory
```

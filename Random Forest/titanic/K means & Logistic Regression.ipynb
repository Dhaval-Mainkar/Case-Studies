# K-means/DBSCAN Clustering

```python
#we do k means just to add another column in our df to check if the score increases
```

```python
cluster_new=DBSCAN(eps=4,min_samples=7) #eps is how many points should the circle cover i.e how many points should points should be there in a cluster and min_Samples is how much cluster/groups must be formed 
cluster_new.fit(x,y)
```

```python
cluster_new.labels_
```

```python
wcss = [] #takes empty array to find points in this array
for i in range(1, 11): #runs 11 times and shift the centroid closer to its cluster
    kmeans = KMeans(n_clusters=i,random_state=0)#i is number of times u want here its 11
    kmeans.fit(x,y)
    wcss.append(kmeans.inertia_) #distance from centroid
```

# do clustering and then plot elbow curve so u will get k accordingly fit it into ur model

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method') # prints graph name
plt.xlabel('Number of clusters')# prints label on x axis
plt.ylabel('WCSS')# prints label on y
plt.show()
```

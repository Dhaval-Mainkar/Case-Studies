```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sp
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("bank-full.csv")
```

```python
df.info()
```

```python
df.isnull().sum() #checkng for null values
```

```python
df
```

```python
df.value_counts("poutcome") #i have done value counts for each column and then selected them for df_new
```

```python
df_new=pd.get_dummies(df,columns=["marital","education","housing","loan","poutcome"])
df_new
```

```python
df_new.columns
```

```python
x=df_new[["marital_single","education_unknown","housing_yes","loan_no","poutcome_other"]]
```

```python
y=df_new["Target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
model.score(x_test,y_test)
```

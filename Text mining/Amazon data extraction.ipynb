# Stars

```python
def get_stars(soup_obj): #here we define get_stars as a class and soup_obj as a object
    try:
        stars= soup_obj.find("a", {"class":"a-link-normal"}).get_text()
        return stars
    except Exception:
        return "nothing"
        print(Exception)
```

```python
print(get_stars(reviews[4]))
```

# Body

```python
def get_review_body(soup_obj)->str:
    try:
        body= soup_obj.find("span",{"class": "a-size-base review-text review-text-content"}).get_text().strip() #strip removes 
                                                                                                                #blank spaces
        return body
    except Exception:
        return "nothing"
        print(Exception)
```

```python
print(get_review_body(reviews[0]))
```

# Date

```python
def date(soup_obj):
    try:
        var= soup_obj.find("span",{"class": "a-size-base a-color-secondary review-date"}).get_text() 
        return var
    except Exception:
        return "nothing"
        print(Exception)
```

```python
print(date(reviews[4]))
```

# Profile name

```python
def name(soup_obj):
    try:
        profile_name= soup_obj.find("span",{"class": "a-profile-name"}).get_text() 
        return profile_name
    except Exception:
        return "nothing"
        print(Exception)
```

```python
print(name(reviews[3]))
```

```python
#calling out all componets in a dictionary
def review_components(soup_obj:BeautifulSoup)->dict:
    return {
        "number of stars":get_stars(soup_obj),
        "date":date(soup_obj),
        "profile_name":name(soup_obj),
        "review":get_review_body(soup_obj)
    }
```

```python
review_components(reviews[0])
```

# Dataframe for one page

```python
records=[review_components(rev) for rev in reviews]
```

```python
df=pd.DataFrame.from_records(records)
df
```

# Data Frame for multiple pages

```python
all_reviews=[]
for p in pages:
    print(p)
    resp = requests.get(p, headers=headers)
    soup = BeautifulSoup(resp.text,"html.parser")
    reviews = soup.find_all("div", {"class": "a-section review aok-relative"})
    records = [review_components(rev) for rev in reviews]
    all_reviews.extend(records)
```

```python
df_final=pd.DataFrame.from_records(all_reviews)
df_final
```

```python
df_final.to_csv("amazon_laptop_reviews.csv")
```

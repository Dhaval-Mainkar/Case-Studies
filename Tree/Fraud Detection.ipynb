# This adds new column called as income where if taxable_income >=30000 then it will be labelled as good or else risky

```python
df_new.loc[df["taxable_income"]>=30000,"income"]="Good"
df_new.loc[df["taxable_income"]<=30000,"income"]="Risky"
```

```python
df_new.head(20)
```

```python
df_new.columns
```

```python
x=df_new[['Undergrad_YES', 'Marital.Status_Divorced', 'Urban_NO']]
y=df_new["income"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.33)
```

```python
for i in range (1,20):
    model=DecisionTreeClassifier(max_depth=i)
    model.fit(x_train,y_train)
    preds=model.predict(x_test)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300) #this just enlarges the image
tree.plot_tree(model);#this will plot
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

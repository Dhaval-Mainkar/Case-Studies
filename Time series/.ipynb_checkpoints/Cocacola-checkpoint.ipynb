# Augumented Dickey Hypothesis

```python
def adf_test(series):
    results=adfuller(series)
    print("Adf statistics:",results[0])
    print("p:",results[1])
    if results[1] <= 0.05:
        print("Data is stationary")
    else:
        print("Data is not stationary")
```

```python
adf_test(df["Sales"])
```

```python
df["first difference"]=df["Sales"]-df["Sales"].shift(1)
```

```python
adf_test(df["first difference"].dropna())
```

```python
df["second difference"]=df["first difference"]-df["first difference"].shift(2)
```

```python
adf_test(df["second difference"].dropna())
```

```python
acf=plot_acf(df["second difference"].dropna()) #p=3
```

```python
pacf=plot_pacf(df["second difference"].dropna()) #q=4
```

# Arima

```python
train=df.head(29)
test=df.tail(13)
```

```python
model=sms.tsa.arima.ARIMA(df["Sales"],order=(3,2,4))
```

```python
arima=model.fit()
```

```python
start=test.index[0]
end=test.index[-1]
```

```python
pred=arima.predict(start,end)
```

```python
test["arima predicted"]=pred
```

```python
test
```

```python
test[["Sales","arima predicted"]].plot()
```

# SARIMAX

```python
df["12th Differencing"]=df["second difference"]-df["second difference"].shift(12)
```

```python
adf_test(df["12th Differencing"].dropna())
```

```python
acf=plot_acf(df["12th Differencing"].dropna())
```

```python
model=SARIMAX(df["Sales"],order=(3,2,4),seasonal=(0,1,0,12))
```

```python
sarimax=model.fit()
```

```python
predict=sarimax.predict(start,end)
```

```python
test["sarimax"]=predict
```

```python
test[["Sales","sarimax","arima predicted"]].plot()
```

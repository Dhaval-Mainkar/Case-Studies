```python
df_train=df_log.head(15)
df_train
```

```python
df_test=df_log.tail(15)
df_test
```

# For train data set

```python
x=df_train["YearsExperience"];y=df_train["Salary"]
```

```python
model=LinearRegression()
```

```python
model.fit(x[:,np.newaxis],y)
```

```python
pred=model.predict(x[:,np.newaxis])
pred
```

```python
score1=model.score(x[:,np.newaxis],y)
```

```python
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.scatter(x,y)
plt.plot(x[:,np.newaxis],pred)
```

# Test data set

```python
a=df_test["YearsExperience"];b=df_test["Salary"]
```

```python
model1=LinearRegression()
```

```python
model1.fit(a[:,np.newaxis],b)
```

```python
pred1=model1.predict(a[:,np.newaxis])#it will predict for b ie:salary
pred1
```

```python
score2=model1.score(a[:,np.newaxis],b)*100
score2
```

```python
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.scatter(a,b)
plt.plot(a[:,np.newaxis],pred1)
```

```python
print("The best score of SLR model for Train dataset using log transformation is:",score1)
print("The best score of SLR model for Test dataset using log transformation is:",score2)
```
